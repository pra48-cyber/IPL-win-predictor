```python
import pandas as pd
import numpy as np
```

```python
match=pd.read_csv("matches.csv")
delivery=pd.read_csv("deliveries.csv")
```

```python
total_score_df=delivery.groupby(['match_id',"inning"]).sum()['total_runs'].reset_index()
```

```python
total_score_df=total_score_df[total_score_df['inning']==1]
```

```python
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python

teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Punjab Kings',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Lucknow Super Giants',
    'Gujarat Titans'
]
```

```python
match_df['team1'].unique()
```

```python
team_name_mapping = {
    'Kings IX Punjab': 'Punjab Kings',
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Rising Pune Supergiant':'Lucknow Super Giants',
    'Rising Pune Supergiants':'Lucknow Super Giants',
    'Gujarat Lions':'Gujarat Titans'
}

```

```python
match_df = match_df.applymap(lambda x: team_name_mapping.get(x, x))
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df
```

```python
delivery_df['total_runs_y'] = pd.to_numeric(delivery_df['total_runs_y'], errors='coerce')

# Apply cumulative sum correctly
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].transform('cumsum')
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna(0)
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x==0 else 1)                                                                   
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)
```

```python
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
```

```python
delivery_df['wickets_left']=10 - wickets
```

```python
delivery_df
```

```python
delivery_df.tail(5)
```

```python
#current run rate=runs/oversd
delivery_df['crr']=delivery_df['current_score']*6/(120-delivery_df['balls_left'])
```

```python
#required run rate
delivery_df['rrr']=(delivery_df['runs_left']*6/delivery_df['balls_left'])
```

```python
delivery_df
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.describe()
```

```python
final_df=final_df[final_df['balls_left']!=0]
final_df
```

```python
final_df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
pipe=Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
